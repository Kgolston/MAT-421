17.2 Linear Interpolation
Linear interpolation is used to find an intermediete point that is on a line between two data points. Using the code from the textbook, *Python Programming and Numerical Methods* I was will analyze a data set for the water levels of lake mead. In 1995, the water level was 1168.91625 feet.

https://www.nps.gov/lake/learn/news/timeline.htm

```python
#Text book example explained w/ a data set from 
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')


x = [1935, 1941, 1956, 1983, 2015]    #Year
y = [708.70, 12220.40, 108.357, 1225.44, 1074.71]   #Water levels in feet

f = interp1d(x, y)
y_hat = f(1995)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1995, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.3 Cubic Spline Interpolation
Cubic spline interpolation is similiar to linear intepolation as it finds a point between two data points. The main difference is the point that is being interpolated is actually a function comprised of a set of piecewise cubic functions.
