20.1 Numerical Differentiation Problem Statement: 
*   A numerical grid is an evenly spaced set of points over the domain of a function
*   The spacing or step size of a numerical grid is the distance between adjacent points on the grid. 
*   A function f(x) can be represented over an interval by computing its value on a grid. Although the function itself may be continuous, this discrete or discretized representation is useful for numerical calculations and corresponds to data sets that may be acquired in engineering and science practice.

```python
#20.2 Finite Difference Approximating Derivatives
#FDA of sin and the derivative cos as the exact solution
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h) 
# compute function
y = np.sin(x)

# compute vector of forward differences
forward_diff = np.diff(y)/h 
# compute corresponding grid
x_diff = x[:-1:] 
# compute exact solution
exact_solution = np.cos(x_diff)   #derivative of y

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between 
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

```python
#20.3 Approximating of Higher Order Derivatives

#Plot the taylor polynomial of degree 3 centered at x=0 for f(x)= (3*exp(x))/(x**2+x+1)
#over the iverval x is an element [-3,3]

#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.misc import derivative

x = np.linspace(-3,3,100)   #domain of x defined
f = lambda x: 3*np.exp(x) / (x**2 + x + 1)    #function defined
y = f(x)
plt.plot(x,y);
plt.show()

#Taylorseries a_n=(f**(n)*(0))/factorial(n) for n=0,1 ,2 ,3
a0 = f(0)
a1 = derivative(f,0,dx=0.001,n=1)
a2 = derivative(f,0,dx=0.001,n=2) / 2
a3 = derivative(f,0,dx=0.001,n=3,order=5) / 6
# The parameter order specifies the number of points to use
# The value order must be odd and at least n + 1

print(a0,a1,a2,a3)

#f(x) and T(x) plot together
T = a0 + a1*x + a2*x**2 + a3*x**3
plt.plot(x,y,x,T), plt.xlim([-3,3]), plt.ylim([0,5]);
plt.show()
```

```python
# 20.4 Numerical Differentiation with Noise
#Computation of sin wave corrupted by a small cosine wave

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01) 
# compute function
omega = 100
epsilon = 0.01

y = np.sin(x) 
y_noise = y + epsilon*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'sin(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()

x = np.arange(0, 2*np.pi, 0.01) 
# compute function
y = np.cos(x) 
y_noise = y + epsilon*omega*np.sin(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of sin(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
